```python
from src.agent.adapters.database import BaseDatabaseAdapter
from src.agent.config import get_database_config
```

```python
db = BaseDatabaseAdapter(get_database_config())

with db as db:
    result = db.execute_query("""SELECT
  COALESCE(SUM(od.quantity * od.final_unit_price), 0) AS total_revenue
FROM
  orders AS o
JOIN
  order_details AS od ON o.order_id = od.order_id
WHERE
  -- Filter for orders shipped today or yesterday
  o.shipped_date >= CURRENT_DATE - INTERVAL '1 year';""")

print(result)
```

## get schema

```python
import sqlalchemy
```

```python
metadata = sqlalchemy.MetaData()
```

```python
metadata.reflect(bind=db._get_connection())
```

```python
for table_name in metadata.tables:
    print(f"- {table_name}")
    for column in metadata.tables[table_name].columns:
        print(f"  - {column.name}")
```
